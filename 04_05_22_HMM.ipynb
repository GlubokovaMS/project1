```python
import pandas as pd
import numpy as np
 
V=np.array((1,3,3,2,3,3,3,2,1,2,1,1,3,1,3,3,3))
print(V)
```

```python
# Transition Probabilities
a = np.array(((0.5, 0.5), (0.4, 0.6)))
 
# Emission Probabilities
b = np.array(((0.5, 0.4, 0.1), (0.2, 0.4, 0.4)))
#b = np.array(((0.5,0.2), (0.4,0.4),(0.1,0.4)))
# Equal Probabilities for the initial distribution
initial_distribution = np.array((0.2, 0.8))
```

```python
def forward(V, a, b, initial_distribution):
    alpha = np.zeros((V.shape[0], a.shape[0]))
    alpha[0, :] = initial_distribution * b[:, V[0]]
 
    for t in range(1, V.shape[0]):
        for j in range(a.shape[0]):
            # Matrix Computation Steps
            #                  ((1x2) . (1x2))      *     (1)
            #                        (1)            *     (1)
            alpha[t, j] = alpha[t - 1].dot(a[:, j])* b[j, V[t]-1]
 
    return alpha
 
alpha = forward(V, a, b, initial_distribution)
print(alpha)
#alpha[t,i]=P("последовательность видимых значений длины t",s(t)=i);i-состояния
```

```python
def forwardlog(V, a, b, initial_distribution):
    alpha = np.zeros((V.shape[0], a.shape[0]))
    alpha[0, :] = np.log(initial_distribution) + np.log(b[:, V[0]])
 
    for t in range(1, V.shape[0]):
        for j in range(a.shape[0]):
            # Matrix Computation Steps
            #                  ((1x2) . (1x2))      *     (1)
            #                        (1)            *     (1)
            alpha[t, j] = np.log(np.exp(alpha[t - 1].dot(a[:, j])+np.log( b[j, V[t]-1])))
            #alpha[t, j] = np.log(np.exp(alpha[t - 1].dot(a[:, j]))* b[j, V[t]-1])
    return alpha
beta = forwardlog(V, a, b, initial_distribution)
#print(beta)
print(np.exp(beta))
```
